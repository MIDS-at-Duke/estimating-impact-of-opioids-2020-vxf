```python
import pandas as pd
import numpy as np
from plotnine import *
```

### Read Data

```python
State_all = pd.read_csv('E:/Study/Duke Class/G2S1/Data Science/Cmder/estimating-impact-of-opioids-2020-vxf/20_intermediate_files/mortality.csv')
```

### Subseting the Desired Reference State

```python
# Identify Reference States 
State_all_subset = State_all[(State_all['State'] == 'WA') |
#Bad shape                             (State_all['State'] == 'MA') |
                             (State_all['State'] == 'CO') |
                             (State_all['State'] == 'UT') |
#Too steep                             (State_all['State'] == 'CA') |
                             (State_all['State'] == 'NC')]
```

### Plotting the State and Confirm the Similarity

```python
(
    ggplot(State_all_subset, aes(x = 'Year', y = 'Deaths', color = 'State')) +
geom_line(alpha = 0.8, size = 0.5)
)
```
