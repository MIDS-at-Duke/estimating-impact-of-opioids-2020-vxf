### For Florida Analaysis, we timespan of interest would be from 2005 to 2015

```python
## subset for year between 2005 and 2015
mortality_state_subset = mortality_state[mortality_state['Year'] >= 2005].copy()
```

```python
mortality_state_subset.head()
```

### Identify reference states first by geographic proximity
* Georgia, South Carolina, Alabama

```python
mortality_state_subset = mortality_state_subset[(mortality_state_subset['State'] == 'FL') | 
                                         (mortality_state_subset['State'] == 'GA') |  
                                         (mortality_state_subset['State'] == 'SC') | 
                                         (mortality_state_subset['State'] == 'AL')
                                        ]
```

```python
mortality_state_subset
```

```python
(
ggplot(mortality_state_subset, aes(x='Year', y='Deaths', color='State')) +
geom_line(alpha = 0.5, size = 0.5)    
)
```

The trend of overdose death was mostly parallel before 2010 where the policy was implemented. 
However, Georgia and South Carolina have way lower death mortality, and it might not be meaningful to compare a high mortality treatment state with low mortality control states.
